```python
from db_connect import get_connection
import pandas as pd

conn = get_connection()
df = pd.read_sql("SELECT TOP 10 * FROM LOV_MATERIAL", conn)
df
```

```python
tables_query = """
SELECT TABLE_SCHEMA, TABLE_NAME
FROM INFORMATION_SCHEMA.TABLES
ORDER BY TABLE_SCHEMA, TABLE_NAME;
"""
tables_df = pd.read_sql(tables_query, conn)
```

```python
views_query = """
SELECT TABLE_SCHEMA, TABLE_NAME
FROM INFORMATION_SCHEMA.VIEWS
ORDER BY TABLE_SCHEMA, TABLE_NAME;
"""
views_df = pd.read_sql(views_query, conn)
```

```python
columns_query = """
SELECT TABLE_NAME, COLUMN_NAME, DATA_TYPE, IS_NULLABLE
FROM INFORMATION_SCHEMA.COLUMNS
ORDER BY TABLE_NAME, ORDINAL_POSITION;
"""
columns_df = pd.read_sql(columns_query, conn)
```

```python
import pandas as pd

with pd.ExcelWriter("db_structure.xlsx") as writer:
    tables_df.to_excel(writer, sheet_name="Tables", index=False)
    views_df.to_excel(writer, sheet_name="Views", index=False)
    columns_df.to_excel(writer, sheet_name="Columns", index=False)

print("Export erfolgreich: db_structure.xlsx erstellt")
```
